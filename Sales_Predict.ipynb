# TASK 5 : SALES PREDICTION USING PYTHON

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"Advertising.csv")
df.head()
```

```python
df.drop(['Unnamed: 0'], axis=1)
```

# Predicting TV Sales

```python
plt.figure(figsize=(16,8))
plt.scatter(
    df['TV'],
    df['Sales'],
    c='black'
)
plt.xlabel('Money spent on TV ads ($)')
plt.ylabel('Sales ($)')
plt.show()
```

```python
X = df['TV'].values.reshape(-1,1)
y = df['Sales'].values.reshape(-1,1)

lr = LinearRegression()
lr.fit(X, y)

print("The linear model is: Y = {:.5} + {:.5}X".format(lr.intercept_[0], lr.coef_[0][0]))
```

```python
predictions = lr.predict(X)

plt.figure(figsize=(16, 8))

plt.scatter(
    df['TV'],
    df['Sales'],
    c='black'
)

plt.plot(
    df['TV'],
    predictions,
    c='green',
    linewidth=2
)

plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Sales ($)")
plt.show()
```

```python
def predict_sales_value(tv_ads_value):
    return 7.0326 + (0.047537 * tv_ads_value)
```

```python
predict_sales_value(230.1)
```

```python
predict_sales_value(150)
```
